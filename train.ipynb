```python
'''
Implementation of a Neural Network as specified in 'Instructions.pdf'. The objective is to semantically segment input
images depicting aerial views of urban environments into the two classes: 'house' and 'not house'. Therefore we currently
only use a single input (rgb.png) and ground-truth image (gt.png). This is done by training a Neural Network.
The task and instructions seem to roughly follow the paper: Fully Convolutional Networks for Semantic Segmentation, so
an FCN is used to solve the problem.

Author: Peter Mller (TU Berlin)
'''

#imports for the neural net
import torch 
import torch.nn as nn
import torch.nn.functional as F

#further imports
from PIL import Image

import matplotlib.pyplot as plt

import numpy as np

#import images
gt_image = Image.open("gt.png")
rgb_image = Image.open("rgb.png")

# Creating numpy-arrays for the processing steps

gt_array = np.array(gt_image).reshape(gt_image.size[0],gt_image.size[1],3)

rgb_array = np.array(rgb_image).reshape(rgb_image.size[0],rgb_image.size[1],4)

'''
Now, the ground-truth image and the training-image are transformed according to certain considerations as explained
along the way. The input is of dimensions: WxHxC, whereas the network-modules expect input of dimensions NxCxHxW.
The necessary transformation is done in the following steps.
'''

# Considering the ground-truth image, assuming that a value of 255 in all 3 color channels means 'white' -> means 
# 'no house'; so we create an array of the same size as the image with its third dimension being 2 
# (same size as the output of the network). A value of 1 means 'corresponding class'; 0 means 'not the corresponding 
# class'

gt_array_trans = np.zeros((gt_array.shape[0],gt_array.shape[1],2))

indizes_1 = np.where(gt_array != 255)
indizes_2 = np.where(gt_array == 255)

gt_array_trans[indizes_1[0],indizes_1[1],0] = 1
gt_array_trans[indizes_2[0],indizes_2[1],1] = 1

gt_array_trans = gt_array_trans[np.newaxis,:,:,:]

gt_array_trans = np.einsum('ijkl->ilkj', gt_array_trans)

# Now, transforming the input image. Also, cutting the last color-channel (black) as required by the Instructions

rgb_array_trans = rgb_array[np.newaxis,:,:,:3]

rgb_array_trans = np.einsum('ijkl->ilkj',rgb_array_trans)

# As a last step we have to cut the images into patches of maximum size 256x256 (WxH) as required by the instructions
# (Also: the authors own RAM is not large enough to train a network on the whole input-image)
# Firstly, we determine how much to cut from each dimension (H and W) to get equal-size patches of 64x64 pixels
# (Everything above that was found to be too memory-consuming on the author's machine). This is assumed to be
# admissible because less than 1% of the image is cropped on each of it's four edges

cut_value_width = int(rgb_array_trans.shape[3] % 64 /2)
cut_value_height = int(rgb_array_trans.shape[2] % 64 /2)

#Now, cropping the images

rgb_array_trans = rgb_array_trans[:,:,cut_value_height:rgb_array_trans.shape[2] -cut_value_height, 
                                  cut_value_width:rgb_array_trans.shape[3] - cut_value_width]

gt_array_trans = gt_array_trans[:,:,cut_value_height:gt_array_trans.shape[2] -cut_value_height, 
                                  cut_value_width:gt_array_trans.shape[3] - cut_value_width]

#Now, cutting the images into 64x64-sized patches an storing these patches into an array representing a mini-batch
# Ordering of patches is: row by row, from left to right in the original input-image dimensions
# Note: A better (more numpy-like) solution seems to be in order here. Due to the time constraint however, the author
# left this part as it is for the time being

number_patches_height = (rgb_array_trans.shape[2]/64)
number_patches_width = (rgb_array_trans.shape[3]/64)

rgb_batch = np.zeros((int(number_patches_height*number_patches_width),3,64,64))
gt_batch = np.zeros((int(number_patches_height*number_patches_width),2,64,64))

patch_counter = 0

for i in range(int(number_patches_height)):
    for j in range(int(number_patches_width)):
        rgb_batch[patch_counter,:,:64,:64] = rgb_array_trans[:,:,i*64:(i+1)*64,j*64:(j+1)*64]
        gt_batch[patch_counter,:,:64,:64] = gt_array_trans[:,:,i*64:(i+1)*64,j*64:(j+1)*64]

        patch_counter += 1
       
    
'''
Here, the network and it's workings are defined.
'''

# input x and expected output y
x = torch.tensor(rgb_batch, dtype = torch.double)
y = torch.tensor(gt_batch, dtype = torch.double)
#x = torch.tensor(rgb_array_trans[:,:,:64,:64], dtype = torch.double)
#y = torch.tensor(gt_array_trans[:,:,:64,:64], dtype = torch.double)

# The neural network is defined using the torch.nn package; layers as described in the task; assuming
# that the upsampling is done bilinearly

model = torch.nn.Sequential(nn.Conv2d(3,16,3,1,1),
                           nn.ReLU(),
                           nn.Conv2d(16,32,3,1,1),
                           nn.ReLU(),
                           nn.MaxPool2d(2,2),
                           nn.Conv2d(32,16,3,1,1),
                           nn.ReLU(),
                           nn.Upsample(scale_factor=2, mode='bilinear'),
                           nn.Conv2d(16,2,5,1,2))
model = model.double()

# Using the mean-squared-error loss-function
loss_fn = torch.nn.MSELoss(size_average=False)

learning_rate = 1e-4
# Using the 'Adam' optimization method
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# Training the network over 2 epoch (meaning in this case that the network is only trained twice on the whole input image
number_epochs = 2

# Keeping track of losses for later plotting
Loss = np.zeros(number_epochs*rgb_batch.shape[0])

for t in range(number_epochs):
    for d in range(rgb_batch.shape[0]):
        # input x and expected output y, training patch by patch
        x = torch.tensor(rgb_batch[d:d+1,:,:,:], dtype = torch.double)
        y = torch.tensor(gt_batch[d:d+1,:,:,:], dtype = torch.double)
        
        # Forward pass: compute predicted y by passing x to the model.
        y_pred = model(x)

        # Compute and print loss.
        loss = loss_fn(y_pred, y)
        print(t, loss.item())
        # Keeping track of the loss
        Loss[(t+1)*d]=loss
    
        # zeroing the gradient before next backward pass
        optimizer.zero_grad()

        # Backward pass: compute gradient of the loss with respect to model parameters
        loss.backward()

        # parameter update
        optimizer.step()
        
        
'''
Plotting the loss over number of image-patches the network is trained on
'''

%matplotlib inline

plt.figure(1)
plt.xlabel='number of image-patches'
plt.ylabel='loss (MSE)'
plt.plot(Loss)
```
